````python
import cv2
from skimage.feature import local_binary_pattern
import matplotlib.pyplot as plt
````

````python
img_path='persona.png'
# Cargar imagen en escala de grises
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)


# Mostrar la imagen original
plt.imshow(img,cmap='gray')
plt.title('Imagen Escala Grises')
plt.axis('off')
plt.show()
````

````python

# Parámetros para LBP
radius = 0  # Radio del círculo
n_points = 0  # Número de puntos en el círculo



# Aplicar el LBP (Local Binary Pattern)
lbp = local_binary_pattern(img ,n_points, radius, method="uniform")

# Mostrar el LBP resultante
plt.figure(figsize=(25, 25))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title(f'Imagen Original {radius}')
plt.axis('off')

plt.subplot(122)
plt.imshow(lbp, cmap='gray')
plt.title('LBP')
plt.axis('off')
plt.show()
````
