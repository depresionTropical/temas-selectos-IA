```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una matriz de 20x20 con 3 canales de color (RGB)
height, width = 20, 20
image = np.zeros((height, width, 3), dtype=np.uint8)

# Definir los colores
background_color = [255, 255, 255]  # Blanco
h_color = [255, 0, 0]  # Rojo

# Rellenar el fondo
image[:, :] = background_color

# Dibujar la letra H
# Columnas de la H
image[4:16, 5:7] = h_color
image[4:16, 12:14] = h_color
# Parte central de la H
image[9:11, 5:14] = h_color

# Visualizar el resultado
plt.imshow(image)
plt.show()
```
